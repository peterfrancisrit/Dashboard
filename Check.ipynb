```python
def cleantype(data):
    clean_data = []
    for i in data.type:
        if 'beans' in i.lower():
            clean_data.append(1)
        elif 'filter' in i.lower():
            clean_data.append(1)
        elif 'espresso' in i.lower():
            clean_data.append(1)
        elif 'blend' in i.lower():
            clean_data.append(1)
        else:
            clean_data.append(0)
    return clean_data
            
    
# data['type_clean'] = data.
```

```python
data['type_clean'] = cleantype(data)
```
